```python
from whoosh import index
from whoosh.fields import Schema, TEXT, NUMERIC
import json
import os
```

```python
with open("../../data/commentary.json", encoding="utf-8") as f:
    commentary = json.load(f)
```

```python
indexing_text = [{"text": commentary[i]["text"], "index": i} for i in range(len(commentary))]
```

```python
# Define schema: 'title' and 'content' fields are text searchable
schema = Schema(text=TEXT(stored=True), index=NUMERIC(stored=True, unique=True))

# Create index directory if it doesn't exist
try:
    DIR = "../../data/bm25"
    if not os.path.exists(DIR):
        os.mkdir(DIR)

    # Create index
    ix = index.create_in(DIR, schema)

    # Open writer to add documents
    writer = ix.writer()

    # Add some documents
    for text in indexing_text:
        writer.add_document(**text)

    writer.commit()

except Exception as e:
    writer.cancel()
    print(f"Failed to write to {DIR} because {e} occur. Writer cancelled.")
```
