```python
from whoosh import index
from whoosh.fields import Schema, TEXT, ID
import os
```

```python
import json
```

```python
with open("../../data/processed/bible.json", encoding="utf-8") as f:
    bible = json.load(f)
```

```python
verses = [{"text": i["text"], "chunk_id": str(i["metadata"]["chunk_id"])} for i in bible]
```

```python
# Define schema: 'title' and 'content' fields are text searchable
schema = Schema(text=TEXT(stored=True), chunk_id=ID(stored=True, unique=True))

# Create index directory if it doesn't exist
DIR = "../../data/processed/bm25"
if not os.path.exists(DIR):
    os.mkdir(DIR)

# Create index
ix = index.create_in(DIR, schema)

# Open writer to add documents
writer = ix.writer()

# Add some documents
for ver in verses:
    writer.add_document(**ver)

writer.commit()
```

```python
writer.cancel()
```
